```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

def gen(m,c):
    x=np.linspace(-5,5,100)
    y=np.linspace(-5,5,100)
    x1,x2=np.meshgrid(x,y)
    p=np.dstack((x1,x2))
    pdf=multivariate_normal(mean=m,cov=c).pdf(p)
    return x1,x2,pdf

distributions=[(np.array([0,0]),np.eye(2)),(np.array([0,0]),np.array([[1,-0.1],[-0.8,1]]))]

fig,ax=plt.subplots(1,2,figsize=(13,5))
plt.suptitle("bivariate normal distribution",color='green')
ax[0].set_title("independent variable")
ax[1].set_title("correlated variable")

for i,d in enumerate(distributions):
    x1,x2,pdf=gen(d[0],d[1])
    con=ax[i].contourf(x1,x2,pdf,100,cmap='magma')
    ax[i].set_xlabel("x1")
    ax[i].set_ylabel("x2")
    ax[i].axis([-3.5,3.5,-3.5,3.5])
    c_bar=fig.colorbar(con,ax=ax[i])
    c_bar.ax.set_ylabel("p(x1,x2)",fontsize=13,color='blue')
```
