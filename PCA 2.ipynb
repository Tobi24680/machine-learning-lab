```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

data=load_breast_cancer()

x=data['data']
scaled_x=StandardScaler().fit_transform(x)
reduced_x=PCA(n_components=2).fit_transform(scaled_x)
plt.figure(figsize=(7,7))
sns.scatterplot(x=np.ndarray.ravel(reduced_x[:,:1]),y=np.ndarray.ravel(reduced_x[:,1:]),hue=data['target'])
plt.title('scatter plot after reducing dimension of feature vectors using pca')
plt.xlabel('feature1')
plt.ylabel('feature2')
plt.grid()
```
